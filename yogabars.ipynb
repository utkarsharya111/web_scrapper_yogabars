```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def cust_name():
    names = soup.find_all('span',class_='a-profile-name')
    cust_name = []
    for j in range(len(names)):
        cust_name.append(names[j].get_text())
    cust_name.pop(0)
    cust_name.pop(0)
    return cust_name 
```

```python
def review_date():
    review_dates = soup.find_all('span',{'data-hook':'review-date'})
    review_date = []
    for j in range(len(review_dates)):
        review_date.append(review_dates[j].get_text())
#     rating.pop(0)
#     rating.pop(0)
    return review_date
```

```python
def rating():
    ratings = soup.find_all('i',class_='review-rating')
    rating = []
    for j in range(len(ratings)):
        rating.append(ratings[j].get_text())
    rating.pop(0)
    rating.pop(0)
    return rating
```

```python
cust_names = []
review_dates = []
ratings = []


for i in range(1,100):
    
    res = requests.get('https://www.amazon.in/Yogabar-Wholegrain-Breakfast-Muesli-Fruits/product-reviews/B07M6KZQCN/ref=cm_cr_arp_mb_paging_btm_2?pageNumber='+str(i))
    soup = BeautifulSoup(res.content,'html.parser')
    
    cust_names.extend(cust_name())
    review_dates.extend(review_date())
    ratings.extend(rating())
```

```python
data = pd.DataFrame({'Customer Name':cust_names,
                     'Review Date':review_dates,
                     'Rating(Out of 5)':ratings})
```
